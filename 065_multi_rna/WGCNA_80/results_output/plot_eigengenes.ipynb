# WGCNA — Cell Type별 분석
- 파라미터 바꿔가며 cell type 하나씩 실행

```python
import os
import scanpy as sc
import PyWGCNA

BASE_DIR = '/data1/project/yeonu/065_multi_rna'
WGCNA_DIR = os.path.join(BASE_DIR, 'WGCNA_80/results_output')
```

---
## 1. Hepatocytes

```python
CT = 'Hepatocytes'
adata = sc.read_h5ad(os.path.join(WGCNA_DIR, f'{CT}_80_pb_n50.h5ad'))

pywgcna = PyWGCNA.WGCNA(
    name=f'{CT}_Network',
    species='homo sapiens',
    anndata=adata,
    TPMcutoff=0,
    networkType='signed hybrid',
    TOMType='signed',
    minModuleSize=30,
    MEDissThres=0.2,
    save=False,
    figureType='png'
)

pywgcna.preprocess(show=False)
pywgcna.findModules()
pywgcna.CalculateSignedKME()
print(f'--- {CT} Done! ---')
```

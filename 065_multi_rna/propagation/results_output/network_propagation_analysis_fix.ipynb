# Network Propagation Analysis — Fixed Cutoff (1/N)

Score cutoff: `1/N` (N = network total genes = 16,201)

균등 분포 대비 높은 확률을 받은 유전자를 선별한다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Hepatocytes

```python
# 1. 결과 읽기
df = pd.read_csv('Hepatocytes_NP_Result.txt', sep='\t', names=['Gene', 'NP_Score'])

# 2. Seed 리스트 읽기
with open('../seed/Hepatocytes_Module1_Cirrhosis_Up_434genes.txt', 'r') as f:
    seed_genes = set([line.strip().split()[0] for line in f if line.strip()])

# 3. Seed 유무 표시
df['is_Seed'] = df['Gene'].apply(lambda x: 'Yes' if x in seed_genes else 'No')

# 4. Rank & P_value 계산
df = df.sort_values(by='NP_Score', ascending=False).reset_index(drop=True)
df['Rank'] = np.arange(1, len(df) + 1)
df['P_value'] = df['Rank'] / len(df)

# 5. 저장
df.to_csv('Hepatocytes_NP_Final_Results_fix.csv', index=False)

# 결과 요약
N = len(df)
SCORE_CUTOFF = 1.0 / N
print(f'Total Genes in Network: {N}')
print(f'Seed Genes found in network: {(df["is_Seed"]=="Yes").sum()} / {len(seed_genes)}')
print(f'Score cutoff (1/N): {SCORE_CUTOFF:.6f}')
print(f'\nTop 20 Genes:')
print(df.head(20).to_string(index=False))
```

```python
df['log_p'] = -np.log10(df['P_value'] + 1e-10)

# Score cutoff: 1/N
N = len(df)
SCORE_CUTOFF = 1.0 / N

# 3-category 분류
color_palette = {'Seed (DEGs)': 'red', 'Above 1/N': 'green', 'Other genes': '#e0e0e0'}
df['Category'] = 'Other genes'
df.loc[df['NP_Score'] >= SCORE_CUTOFF, 'Category'] = 'Above 1/N'
df.loc[df['is_Seed'] == 'Yes', 'Category'] = 'Seed (DEGs)'

plt.figure(figsize=(10, 8))
sns.set_style('whitegrid')
for cat in ['Other genes', 'Above 1/N', 'Seed (DEGs)']:
    sub = df[df['Category'] == cat]
    plt.scatter(sub['log_p'], sub['NP_Score'], c=color_palette[cat],
                label=cat, alpha=0.6, s=50)

plt.axhline(y=SCORE_CUTOFF, color='black', linestyle='--',
            label=f'1/N = {SCORE_CUTOFF:.6f}')

# Top 10 non-seed 라벨링
top_non_seed = df[df['is_Seed'] == 'No'].nlargest(10, 'NP_Score')
for _, row in top_non_seed.iterrows():
    plt.text(row['log_p'], row['NP_Score'], row['Gene'],
             fontsize=9, fontweight='bold')

plt.title('Hepatocytes — Network Propagation (cutoff = 1/N)', fontsize=14)
plt.xlabel('-log10(p-value)', fontsize=12)
plt.ylabel('NP Score', fontsize=12)
plt.legend()
plt.tight_layout()
plt.xlim(0, 4.4)
plt.ylim(0, 0.0083)
plt.show()

# Summary
candidates = df[(df['NP_Score'] >= SCORE_CUTOFF) & (df['is_Seed'] == 'No')]
print(f'SCORE_CUTOFF (1/N): {SCORE_CUTOFF:.6f}')
print(f'Above 1/N (non-seed): {len(candidates)}')
print(f'\n=== Novel Candidate Genes ({len(candidates)}) ===')
if len(candidates) > 0:
    print(candidates[['Gene', 'NP_Score', 'P_value', 'Rank']].to_string(index=False))
else:
    print('No novel candidate genes found.')
```
